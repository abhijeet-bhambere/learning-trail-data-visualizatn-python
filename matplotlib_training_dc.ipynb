# Small Multiples
Showing similar plots across multiple conditions -- we pass a tuple to the `sublplots` for adding  _X by Y grid_ of sub-plots


```python
# Adding 2 X 1 grid 
fig, ax = plt.subplots(2,1)

# Adding plot for Austin
ax[0].plot(austin["MONTH"],austin["MLY-PRCP-NORMAL"],marker='o',color="r")
ax[0].plot(austin["MONTH"],austin["MLY-PRCP-25PCTL"],marker='o',color="r",linestyle='--')
ax[0].plot(austin["MONTH"],austin["MLY-PRCP-75PCTL"],marker='o',color="r",linestyle='--')

# Adding plot for Seattle
ax[1].plot(seattle["MONTH"],seattle["MLY-PRCP-NORMAL"],marker='v',color="purple")
ax[1].plot(seattle["MONTH"],seattle["MLY-PRCP-25PCTL"],marker='v',color="purple",linestyle='--')
ax[1].plot(seattle["MONTH"],seattle["MLY-PRCP-75PCTL"],marker='v',color="purple",linestyle='--')

ax[0].set_xlabel("Months")
ax[0].set_ylabel("Rainfall precipitate (in)")
# Adding title to top-most subplot 
ax[0].set_title("Average monthly rainfall")

plt.figure(figsize=(20,6))
plt.show()
```

```python
# Adding 2 X 1 grid 
fig, ax = plt.subplots(2,1,sharey=True)

# Adding plot for Austin
ax[0].plot(austin["MONTH"],austin["MLY-PRCP-NORMAL"],marker='o',color="r")
ax[0].plot(austin["MONTH"],austin["MLY-PRCP-25PCTL"],marker='o',color="r",linestyle='--')
ax[0].plot(austin["MONTH"],austin["MLY-PRCP-75PCTL"],marker='o',color="r",linestyle='--')

# Adding plot for Seattle
ax[1].plot(seattle["MONTH"],seattle["MLY-PRCP-NORMAL"],marker='v',color="purple")
ax[1].plot(seattle["MONTH"],seattle["MLY-PRCP-25PCTL"],marker='v',color="purple",linestyle='--')
ax[1].plot(seattle["MONTH"],seattle["MLY-PRCP-75PCTL"],marker='v',color="purple",linestyle='--')

# Adding x-asix label to bottom plot
ax[1].set_xlabel("Months")
ax[0].set_ylabel("Rainfall precipitate (in)")
ax[1].set_ylabel("Rainfall precipitate (in)")
# Adding title to top-most subplot 
ax[0].set_title("Average monthly rainfall")

plt.figure(figsize=(20,6))
plt.show()
```

```python
# Creating a 2X2 grid -- plotting prcp in left columns & temperatures in right columns
fig, ax = plt.subplots(2, 2,figsize=(14,8))

# Creating shared y-axis for respective plots
ax[0,1].sharey(ax[1,1])
ax[0,0].sharey(ax[1,0])

# Adding plots for Austin
ax[0,0].plot(austin["MONTH"],austin["MLY-PRCP-NORMAL"],marker='o',color="r")
ax[0,1].plot(austin["MONTH"],austin["MLY-TAVG-NORMAL"],marker='1',color="r")
# Adding plots for Seattle
ax[1,0].plot(seattle["MONTH"],seattle["MLY-PRCP-NORMAL"],marker='o',color="purple")
ax[1,1].plot(seattle["MONTH"],seattle["MLY-TAVG-NORMAL"],marker='1',color="purple")

# Adding titles
ax[0,0].set_title("Austin Avg Rainfall")
ax[0,1].set_title("Austin Avg Temp")

ax[1,0].set_title("Seattle Avg Rainfall")
ax[1,1].set_title("Seattle Avg Temp")

ax[0,0].set_ylabel("Rainfall precipitate (in)")
ax[1,0].set_ylabel("Rainfall precipitate (in)")

plt.show()
```

# Plotting time-series for two different scales
There'll be 2 different y-axis variables. This is done using the `.twinx()` method

```python
# Also added plot line colors to better differentiate both lines
fig, ax = plt.subplots()
# Creating the plot for columnn-1
ax.plot(co2_data.index,co2_data["co2"],color='r')
ax.set_xlabel("Time")
ax.set_ylabel("CO2 (ppm)")

# Creating a twin axis for column-2
ax2 = ax.twinx()
ax2.plot(co2_data.index,co2_data["relative_temp"],color='b')
# Mention the 
ax2.set_ylabel("Deg C")

plt.show()
```

## Adding ticks & colors
for both ticks (`ax.tick_params`) and axis labels (`set_xlabels(...,color=)`)

```python
# Also added plot line colors to better differentiate both lines
fig, ax = plt.subplots()
# Creating the plot for columnn-1
ax.plot(co2_data.index,co2_data["co2"],color='#dc2f02')
ax.set_xlabel("Time")
ax.set_ylabel("CO2 (ppm)",color='#dc2f02')
# Setting the ticks colors
ax.tick_params('y',colors='#dc2f02')

# Creating a twin axis for column-2
ax2 = ax.twinx()
ax2.plot(co2_data.index,co2_data["relative_temp"],color="#390099")
ax2.set_ylabel("Deg C",color="#390099")
ax2.tick_params('y',colors="#390099")

plt.show()
```

# Defining a function for styling
So the parameters can be re-used for every other plot. We'll only need to pass in the arguments

```python
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
    '''Set the a, x-axis, y-axis, color of tickets & also the text on x & y-axis labels '''
    axes.plot(x,y,color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel,color=color)
    axes.tick_params('y',colors=color)
```

```python
fig, ax = plt.subplots(figsize =(10,6))
plot_timeseries(axes=ax,x=co2_data.index,y=co2_data["co2"],color='#dc2f02',xlabel="Time",ylabel="CO2(ppm)")

ax2=ax.twinx()
plot_timeseries(axes=ax2,x=co2_data.index,y=co2_data["relative_temp"],color='#390099',xlabel="Time",ylabel="Relative temp")


plt.show()
```

# Adding Annotations
### `axes.annotate(text= ,xy= ,xytext= ,arrowprops= )`
 - `text` is for the annotation label to be displayed
 - `xy` decides the point to be annotated
 - `xytext` is to place the location of the label box
 - `arrowprops` is for formatting the linestyle of the arrow mark

```python
# Finding dates with the co2 minimum and maximum values
co2_min = co2_data["co2"].min()
co2_max = co2_data["co2"].max()

co2_data[(co2_data["co2"]==co2_min) | (co2_data["co2"]==co2_max)]
```

```python
rt_min = co2_data["relative_temp"].min()
rt_max = co2_data["relative_temp"].max()

co2_data[(co2_data["relative_temp"]==rt_min)]
```

```python
fig, ax = plt.subplots(figsize =(12,6))
plot_timeseries(axes=ax,x=co2_data.index,y=co2_data["co2"],color='#dc2f02',xlabel="Time",ylabel="CO2(ppm)")

ax2=ax.twinx()
plot_timeseries(axes=ax2,x=co2_data.index,y=co2_data["relative_temp"],color='#390099',xlabel="Time",ylabel="Relative temp")

ax.annotate(text="CO2 min",xy=(pd.Timestamp("1958-09-06"),313.20),xytext=(pd.Timestamp("1960-01-01"),360),arrowprops={"arrowstyle":"->","color":"black"})

plt.show()
```

```python
# Customizing the annotation even further
fig, ax = plt.subplots(figsize =(12,6))
plot_timeseries(axes=ax,x=co2_data.index,y=co2_data["co2"],color='#dc2f02',xlabel="Time",ylabel="CO2(ppm)")

ax2=ax.twinx()
plot_timeseries(axes=ax2,x=co2_data.index,y=co2_data["relative_temp"],color='#390099',xlabel="Time",ylabel="Relative temp")

arrowprops=dict(arrowstyle="->",linestyle='--',color="black",linewidth=1,connectionstyle='angle,angleA=-70,angleB=70')
bbox=dict(facecolor='yellow',boxstyle='round')

ax.annotate(text="CO2 min",xy=(pd.Timestamp("1958-09-06"),313.20),xytext=(pd.Timestamp("1960-01-01"),360),arrowprops=arrowprops,bbox=bbox)

plt.show()
```

# Quantitative comparisons: bar-charts
- Comparing parts of the data within itself
- For example, a stacked bar chart can show the total of a category while also giving idea of constituent proportions

```python
medals = pd.read_csv("course_datasets/medals_by_country_2016.csv",index_col=0)
medals
```

```python
fig,ax = plt.subplots(figsize=(7,3))
ax.bar(medals.index,medals["Gold"])

ax.set_xlabel("Country")
ax.set_ylabel("Medals tally")


plt.show()
```

```python
# Rotating the x-axis ticks to remove the over-lapping of country names 
fig,ax = plt.subplots()
ax.bar(medals.index,medals["Gold"])

# ax.set_xlabel("Country")
ax.set_xticklabels(medals.index,rotation=90)
ax.set_ylabel("Medals tally")


plt.show()
```

```python
# Adding visualization of other medals categories
fig,ax = plt.subplots()
ax.bar(medals.index,medals["Gold"])
# Adding the Silver medals  -- creating a Stacked bar chart
ax.bar(medals.index,medals["Silver"],bottom=medals["Gold"])

# Rotating the x-axis ticks to remove the over-lapping of country names 
ax.set_xticklabels(medals.index,rotation=90)
ax.set_ylabel("Medals tally")

ax.legend(loc="upper right",title="legend")
plt.show()
```

```python
# Adding visualization of other medals categories
fig,ax = plt.subplots()
ax.bar(medals.index,medals["Gold"] , label="Gold")
# Adding the Silver medals  -- creating a Stacked bar chart
ax.bar(medals.index,medals["Silver"],bottom=medals["Gold"] , label="Silver")
# Also adding the Bronze medals to the stack
ax.bar(medals.index,medals["Bronze"],bottom=medals["Gold"] + medals["Silver"] , label="Bronze")

# Rotating the x-axis ticks to remove the over-lapping of country names 
ax.set_xticklabels(medals.index,rotation=90)
ax.set_ylabel("Medals tally")

# Adding a legend
ax.legend()
plt.show()
```

# Quantitative comparisons: histograms
Helps understand distribution of values in a dataset

```python
players = pd.read_csv("course_datasets/summer2016.csv",index_col=0)
players.head()
```

```python
players["Medal"].value_counts()
```

#### We'll create separate dataframes for men's medal winners in Rowing & ymnastics 
We'll observe the distribution of various  in these respective sports & how one sport's medal winners compare to the medal winners of the other sport

```python
# Creating separate dataframes for two sports
gymnastics = players[players["Sport"]=="Gymnastics"]
gymnastics.head()

rowing = players[players["Sport"]=="Rowing"]
rowing.head()
```

```python
# Plotting heights of each category
fig, ax = plt.subplots()
ax.bar("Rowing",rowing["Height"].mean(), label = "Rowing")
ax.bar("Gymnastics",gymnastics["Height"].mean(), label = "Gymnastics")

ax.set_ylabel("Avgerage player height (cm)")
ax.set_title("Player heights")

ax.legend()
plt.show()
```

### While the above plot does give an idea that average height of a Rowing player is more than that of a Gymnastics player
A **histogram** will give better idea of how the individual values (heights) are distributed in both the categories(rowing & gymnastics)

```python
# Plotting heights of each category
fig, ax = plt.subplots()
ax.hist(rowing["Height"],label="Rowing")
ax.hist(gymnastics["Height"],label="Gymnastics")

ax.set_xlabel("Player height (cm)")
ax.set_ylabel("# of observations")
ax.set_title("Distribution of player heights")

ax.legend()

plt.show()
```

```python
# We can increase or decrease the granularity of the histogram by changing the no. of bins
# These are the buckets into which the counts can be categorized into
fig, ax = plt.subplots()
ax.hist(rowing["Height"],label="Rowing",bins=20)
ax.hist(gymnastics["Height"],label="Gymnastics",bins=20)

ax.set_xlabel("Player height (cm)")
ax.set_ylabel("# of observations")
ax.set_title("Distribution of player heights")

ax.legend()

plt.show()
```

### Observations:
From looking at above distribution, we observe that:
- There are approx. 16-18 tall players in the Gymnastics team
- There are approx 30 tall players in the Rowing team in 180's and 190's cm each

```python
# We can also format the histogram 
fig, ax = plt.subplots()
ax.hist(rowing["Height"],label="Rowing",bins=20,histtype='step')
ax.hist(gymnastics["Height"],label="Gymnastics",bins=20,histtype='step')

ax.set_xlabel("Player height (cm)")
ax.set_ylabel("# of observations")
ax.set_title("Distribution of player heights")

ax.legend()

plt.show()
```

# Scatter plots
Used to understand correlation between two variables across observations. Also called bivariate comparison.

```python
fig, ax = plt.subplots(figsize=(8,6))
# Plotting co2 v/s relative temp data
ax.scatter(co2_data["co2"],co2_data["relative_temp"])

ax.set_xlabel("CO2(ppm)")
ax.set_ylabel("Relative temp(degC)")


plt.show()
```

## Showing two bivariate comarisons side-by-side

```python
# Showing two bivariate comarisons side-by-side
eighties = co2_data["1980-01-06":"1989-12-31"]
nineties = co2_data["1990-01-06":"1999-12-31"]
```

```python
fig, ax = plt.subplots()

ax.scatter(eighties["co2"],eighties["relative_temp"],label="eighties")
ax.scatter(nineties["co2"],nineties["relative_temp"],label="nineties")

ax.set_xlabel("CO2(ppm)")
ax.set_ylabel("Relative temp(degC)")
ax.set_title("CO2 vs Relative Temperatures")

ax.legend()
plt.show()
```

Thus we see that relation between CO2 and relative temperatures didnt change much during the 1980s. However relative temperature increased with an increase in CO2 levels during the 1990s

```python
# A third variable can also be added to the plot that can be denoted by a color gradient of the points
fig, ax = plt.subplots(figsize=(8,6))
# Plotting co2 v/s relative temp data
ax.scatter(co2_data["co2"],co2_data["relative_temp"],c=co2_data.index)

ax.set_xlabel("CO2(ppm)")
ax.set_ylabel("Relative temp(degC)")

plt.show()
```

```python
# The 'c' parameter can be used to indicate values of any of the variables
fig, ax = plt.subplots(figsize=(8,6))

# You can add other color gradients too
# Plotting co2 v/s relative temp data
ax.scatter(co2_data["co2"],co2_data["relative_temp"],c=co2_data["relative_temp"],cmap="OrRd")

ax.set_xlabel("CO2(ppm)")
ax.set_ylabel("Relative temp(degC)")


plt.show()
```
